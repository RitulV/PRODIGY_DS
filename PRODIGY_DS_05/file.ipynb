<h5>As there are almost 46 columns in the dataset, we will try to split the dataset to better analyse it</h5>

```python
df_num = df.select_dtypes(exclude = ['object'])
df_obj = df.select_dtypes(include = ['object'])
print(len(df_num.columns))
print(len(df_obj.columns))
```

<h4>Cleaning the 'df_num' column</h4>

```python
# Calculating the missing values in 'df_num' dataframe

df_m = df_num.isnull().sum(axis=0).reset_index()
df_m.columns = ['column', 'null_count']
df_m[df_m['null_count']>0].sort_values(by='null_count')
```

```python
# removing columns with major missing values

df_num.drop(labels=['Wind_Chill(F)', 'Precipitation(in)', 'End_Lat', 'End_Lng','Turning_Loop'], axis=1, inplace=True)

# also removing the date and time column

df_num = df_num.select_dtypes(exclude = ['object', 'datetime64'])
```

<h4>Cleaning the 'df_obj' column</h4>

```python
# Calculating the missing values in 'df_obj' dataframe

df_m = df_obj.isnull().sum(axis=0).reset_index()
df_m.columns = ['column', 'null_count']
df_m[df_m['null_count']>0].sort_values(by='null_count')
```

```python
# removing columns with major missing values

df_obj.drop(labels=['County', 'Description'], axis=1, inplace=True)

# also removing the date and time column

df_obj = df_obj.select_dtypes(include = ['object'])
```

<h5>Combining both the dataframes</h5>

```python
df_clean = pd.concat([df_num,df_obj],axis = 1)
df_clean.drop(columns=['Country'], inplace=True)
```

```python
df_clean.columns
```

<h5>State-wise data distribution</h5>

```python
# Getting the data for state and accidents

state_acc_counts = pd.DataFrame(df_clean['State'].value_counts())
# state_acc_counts
x = state_acc_counts.index.to_list()
```

```python
fig,axs = plt.subplots(figsize = (10,6))

x = state_acc_counts[0:15].index.to_list()
y = state_acc_counts[0:15].values.flatten()

sns.barplot(x=x, y = y)
axs.tick_params(axis = 'x', rotation = 90)
axs.set_ylabel("Accidents")
axs.set_xlabel("State")
plt.title("Top 15 States with Highest Number of Accidents")

plt.show()
```

<h5>City-wise data distribution</h5>

```python
city_acc_counts = pd.DataFrame(df_clean['City'].value_counts()).reset_index()
# city_acc_counts
city_acc_counts.columns = ['City',"Number of Accidents"]
city_acc_counts.sort_values(by = 'Number of Accidents', ascending = False, inplace = True)
x = city_acc_counts['City'][:15].to_list()
y = city_acc_counts["Number of Accidents"][:15]
```

```python
fig,axs = plt.subplots(figsize = (10,6))

sns.barplot(x = x, y = y)
axs.tick_params(axis = 'x', rotation = 90)
axs.set_ylabel("Number of Accidents")
axs.set_xlabel("Cities")
plt.title("Top 15 Cities with Highest Number of Accidents")

plt.show()
```
